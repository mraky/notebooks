```python
from datetime import datetime
from elasticsearch import Elasticsearch
import pandas as pd
from mjupyutils import show
import json

es = Elasticsearch(['http://10.13.1.150:9200'])
res = es.search(size=1600,index="sledovani", body={"query": {"match_all": {}}})
print("Got %d Hits:" % res['hits']['total'])
aa=[]
for hit in res['hits']['hits']:
    stamp=hit['_id']
    lat=hit['_source']['latitude']
    long=hit['_source']['longtitude']
    #ship=hit['_source']['lod']
    #print("stamp: {st}  latitude: {la} longtitude:{lo}".format(st=stamp,la=lat,lo=long))
    aa.append([stamp,lat,long])
    
df = pd.DataFrame(aa)
show.table(df[[1,2]])
show.map(df[[1,2]].to_json(orient="values"))
```
